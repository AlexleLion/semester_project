```python
import os
import pandas as pd
from collections import Counter
from wordcloud import WordCloud

import matplotlib.pyplot as plt

# Fonction pour charger les fichiers d'occurrences de mots
def load_word_occurrences(directory):
    word_occurrences = {}
    for filename in os.listdir(directory):
        if filename.endswith(".csv"):
            filepath = os.path.join(directory, filename)
            df = pd.read_csv(filepath)
            word_occurrences[filename] = Counter(dict(zip(df['Mot'], df['Occurences'])))
    return word_occurrences

# Fonction pour analyser les occurrences de mots
def analyze_word_occurrences(word_occurrences):
    total_occurrences = Counter()
    for occurrences in word_occurrences.values():
        occurrences.subtract(Counter({'femme': occurrences['femme'], 'enfant': occurrences['enfant'], 'homme': occurrences['homme']}))
        total_occurrences.update(occurrences)
    return total_occurrences

# Fonction pour visualiser les r√©sultats
def visualize_word_occurrences(word_occurrences):
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_occurrences)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()
```

```python
# Charger les fichiers d'occurrences de mots
directory = '../data/occurences/ouvrages_mixtes/'  
word_occurrences = load_word_occurrences(directory)

# Analyser les occurrences de mots
total_occurrences = analyze_word_occurrences(word_occurrences)

# Visualiser les r√©sultats
visualize_word_occurrences(total_occurrences)
```
