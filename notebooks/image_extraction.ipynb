## essai 1


```python
import cv2
import os

# Charger l'image
image_path = "../sources/ouvrages_mixtes/01/page_60.jpg"
image = cv2.imread(image_path)
```

```python

# Convertir en niveaux de gris
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Appliquer un seuil binaire
_, binary = cv2.threshold(gray, 200, 255, cv2.THRESH_BINARY_INV)

# Trouver les contours
contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Trouver le plus grand contour (la partie graphique centrale)
largest_contour = max(contours, key=cv2.contourArea)

# Extraire les coordonnées du rectangle englobant
x, y, w, h = cv2.boundingRect(largest_contour)

# Extraire la partie graphique
extracted_graphic = image[y:y+h, x:x+w]

# Sauvegarder l'image extraite
output_path = "../images/ouvrages_mixtes/extracted_graphic.jpg"
os.makedirs(os.path.dirname(output_path), exist_ok=True)
cv2.imwrite(output_path, extracted_graphic)

print(f"Partie graphique extraite et sauvegardée sous : {output_path}")
```

## Essai 2


```python
import fitz  # PyMuPDF

def extract_images_from_pdf(pdf_path, output_folder):
    # Ouvrir le fichier PDF
    pdf_document = fitz.open(pdf_path)
    image_count = 0
    os.makedirs(output_folder, exist_ok=True)

    for page_number in range(len(pdf_document)):
        page = pdf_document[page_number]
        images = page.get_images(full=True)  # Obtenir toutes les images de la page

        for img_index, img in enumerate(images):
            xref = img[0]
            # Extraire l'image
            base_image = pdf_document.extract_image(xref)
            image_bytes = base_image["image"]
            image_extension = base_image["ext"]  # Format de l'image (png, jpeg, etc.)

            # Sauvegarder l'image
            image_output_path = f"{output_folder}/page_{page_number + 1}_img_{img_index + 1}.{image_extension}"
            with open(image_output_path, "wb") as image_file:
                image_file.write(image_bytes)

            image_count += 1
            print(f"Image sauvegardée : {image_output_path}")

    print(f"Extraction terminée ! {image_count} images extraites.")
    pdf_document.close()
```

```python

# Chemin du fichier PDF et dossier de sortie
pdf_path = "../sources/ouvrages_mixtes/01.pdf"  # Remplacez par le chemin vers votre fichier PDF
output_folder = "../images/ouvrages_mixtes"  # Dossier de sortie

# Appeler la fonction
extract_images_from_pdf(pdf_path, output_folder)
```
